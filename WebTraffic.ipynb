```python
import io
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from google.colab import files

# ==========================================
# 1. FILE UPLOAD PROMPT
# ==========================================
print("Click the button below to upload your web traffic dataset.")
uploaded = files.upload()

# Get the filename and load the data
filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]))

# Ensure the column names match perfectly (stripping spaces)
df.columns = df.columns.str.strip()

print(f"\nSuccessfully loaded '{filename}'!")
print(f"Detected Columns: {list(df.columns)}\n")

# ==========================================
# 2. TRAFFIC OVERVIEW & ENGAGEMENT
# ==========================================
print("=== [1] CORE WEB TRAFFIC METRICS ===")
# Calculate overall averages from your metrics
total_views = df['Page Views'].sum() if 'Page Views' in df.columns else "N/A"
avg_duration = df['Session Duration'].mean() if 'Session Duration' in df.columns else "N/A"
avg_bounce = df['Bounce Rate'].mean() if 'Bounce Rate' in df.columns else "N/A"
avg_conversion = df['Conversion Rate'].mean() if 'Conversion Rate' in df.columns else "N/A"

print(f"🔹 Total Page Views:          {total_views:,}")
print(f"🔹 Average Session Duration:  {avg_duration:.2f} minutes/seconds")
print(f"🔹 Average Bounce Rate:      {avg_bounce:.2%}" if isinstance(avg_bounce, float) and avg_bounce <= 1 else f"🔹 Average Bounce Rate:      {avg_bounce:.2f}%")
print(f"🔹 Average Conversion Rate:  {avg_conversion:.2%}" if isinstance(avg_conversion, float) and avg_conversion <= 1 else f"🔹 Average Conversion Rate:  {avg_conversion:.2f}%")
print("\n" + "="*40 + "\n")

# ==========================================
# 3. TRAFFIC SOURCE ANALYSIS
# ==========================================
if 'Traffic Source' in df.columns:
    print("=== [2] TRAFFIC PERFORMANCE BY SOURCE ===")
    # Grouping performance metrics by how users arrived
    source_summary = df.groupby('Traffic Source').agg({
        'Page Views': 'sum',
        'Session Duration': 'mean',
        'Conversion Rate': 'mean'
    }).sort_values(by='Page Views', ascending=False)

    display(source_summary)
    print("\n" + "="*40 + "\n")

# ==========================================
# 4. IDENTIFYING DROP-OFF & RETENTION
# ==========================================
print("=== [3] HIGH DROP-OFF / BOUNCE ANALYSIS ===")
if 'Bounce Rate' in df.columns:
    # Sort to find pages or sources with the highest bounce/drop-off rate
    high_dropoff = df.sort_values(by='Bounce Rate', ascending=False).head(5)
    print("Top 5 Areas with the Highest Drop-off (Bounce Rate):")
    display(high_dropoff)
else:
    print("Bounce Rate column missing or improperly formatted.")
print("\n" + "="*40 + "\n")

# ==========================================
# 5. INTERACTIVE PERFORMANCE VISUALIZATION
# ==========================================
print("=== [4] TRAFFIC VS CONVERSION VISUALIZATION ===")
# Creates a scatter map analyzing traffic volume vs how well it converts

# Check if we can map by Traffic Source or look at overall rows
color_param = 'Traffic Source' if 'Traffic Source' in df.columns else None

fig = px.scatter(
    df,
    x='Page Views',
    y='Conversion Rate',
    size='Session Duration' if 'Session Duration' in df.columns else None,
    color=color_param,
    hover_data=['Bounce Rate', 'Time on Page'] if 'Time on Page' in df.columns else ['Bounce Rate'],
    title="Traffic Volume vs. Conversion Rate (Size = Session Duration)",
    template="plotly_dark"
)

fig.update_layout(height=500)
fig.show()
```
